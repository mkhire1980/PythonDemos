```
import matplotlib.pyplot as plt
import numpy as np

years = np.array(['2012','2013','2014','2015'])
sales_africa = np.array(['23456.12','34567.34','45678.45','56789.56'])
sales_usa = np.array(['134561.12','234569.34','348678.45','256789.56'])
sales_latam = np.array(['234563.12','234767.34','145678.45','356789.56'])
sales_asia = np.array(['345634.12','234507.34','845678.45','556789.56'])
sales_europe = np.array(['134567.12','434267.34','105678.45','656789.56'])

fig,ax = plt.subplots()
europe, = ax.plot(years,sales_europe);
europe.set_label("Europe")

usa, = ax.plot(years,sales_usa);
usa.set_label("USA")
usa.set_dashes([2,2,2,2])

africa, = ax.plot(years,sales_africa);
africa.set_label("Africa")
africa.set_dashes([2,2,10,2])

asia, = ax.plot(years,sales_asia);
asia.set_label("Asia")
asia.set_dashes([2,2,15,2])

latam, = ax.plot(years,sales_latam);
latam.set_label("Latam")
latam.set_dashes([2,2,1,2])

plt.legend()
```

```
fig,ax = plt.subplots(ncols=3, nrows=2, sharex = True, sharey=True)
europe, = ax[0][0].plot(years,sales_europe)
europe.set_label("Europe")
usa = ax[0][1].bar(years,sales_usa)
usa.set_label("USA")
asia = ax[0][2].scatter(years,sales_asia)
asia.set_label("ASIA")

africa, = ax[1][0].plot(years,sales_africa)
africa.set_label("Africa")
africa.set_dashes([2,2,10,2])
```

```
fig,ax = plt.subplots(ncols=4, nrows=1, sharex = True, sharey=True)
europe, = ax[0].plot(years,sales_europe)
europe.set_label("Europe")
ax[0].set_title("sales in Europe")

usa = ax[1].bar(years,sales_usa)
usa.set_label("USA")
ax[1].set_title("sales in USA")

asia = ax[2].scatter(years,sales_asia)
asia.set_label("ASIA")
ax[2].set_title("sales in ASIA")

africa = ax[3].bar(years,sales_africa,width=0.5, color="red")
latam = ax[3].bar(years,sales_latam,width=0.5, color="seagreen");
ax[3].set_title("sales in Africa and LATAM")
ax[3].legend((asia,latam),('Africa','Asia'))

fig.set_size_inches(20.5,5.5, forward=True)
```
